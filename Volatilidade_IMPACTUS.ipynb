```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web

data_inicio = ''
data_final = ''
ticker = ''
gr치fico = 'd' # d = day; w = week; m = month

cotacoes_df = web.get_data_yahoo('{}'.format(ticker), start='{}'.format(data_inicio), end='{}'.format(data_final), interval='{}'.format(gr치fico))

# Retornos
cotacoes_df['Retornos'] = cotacoes_df['Adj Close'].pct_change()

# Retornos Acumulados
cotacoes_df['Retorno Acumulado {}'.format(ticker)] = ((1 + cotacoes_df['Retornos']).cumprod()) # coluna de retornos acumulados
cotacoes_df['Retorno Acumulado {}'.format(ticker)].iloc[0] = 1 # coluna de retornos acumulados
display(cotacoes_df)

#calculo volatilidade di치ria
volatilidade_diaria = np.std(cotacoes_df['Retornos'])
print('Volatilidade Di치ria: {:.2%}'.format(volatilidade_diaria))

#calculo volatilidade anual
volatilidade_anual = (volatilidade_diaria * np.sqrt(252))
print('Volatilidade Anual: {:.2%}'.format(volatilidade_anual))
```
